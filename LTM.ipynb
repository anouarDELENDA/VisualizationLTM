```python

def init_fig():
    #fig = plt.figure()
    #fig = go.FigureWidget()
    points = {"x": [0, 3, 0, 0],
              "y": [0, 0, 3, 0],
              "z": [0, 0, 0, 3]}

    points_vert = {"x": [0, 3, 0, 0, 0, 0, 0, 3],
                   "y": [0, 0, 3, 0, 0, 3, 0, 0],
                   "z": [0, 0, 0, 3, 0, 0, 3, 0]}
    
    #ax = fig.add_subplot(projection='3d')          
    #ax.scatter(xs = points["x"], ys= points["y"], zs= points["z"])
    layout = go.Layout(title='Linear Tetrahedron Method', 
                    barmode='overlay', 
                    xaxis={}, 
                    yaxis={})

    trace1 = go.Scatter3d(
            # 4 points of a tetrahedron
            x= points["x"],
            y= points["y"],
            z=points["z"],
            mode='markers+text',
            )
    trace2 = go.Scatter3d(
              # 4 vertices of a tetrahedron
                x= points_vert["x"],
                y= points_vert["y"],
                z= points_vert["z"],
                line=dict(
                    color='black',
                    width=2,
                ),
                hoverinfo='skip',
                mode='lines',
            )
    data = [trace1, trace2]
    
    
    
    
    #fig.add_trace(
    #    go.Scatter3d(
    #        # 4 vertices of a tetrahedron
    #        x= points["x"],
    #        y= points["y"],
    #        z=points["z"],
    #        mode='markers+text',
    #        ))
     
    #ax.plot(xs = points_vert["x"], ys= points_vert["y"], zs= points_vert["z"])

    #df_pts = pd.DataFrame(points)

    #for (p_name, (i, p)) in zip([r"$E_1$", r"$E_2$", r"$E_3$", r"$E_4$"], enumerate(df_pts.iterrows())) :
    #    ax.text(p[1][0], p[1][1], p[1][2], p_name, color='red') 
    #    #ax.plot(xs = df_pts["x"][i:i+1], ys= df_pts["y"][i:i+1], zs= df_pts["z"][i:i+1])  
#
    #ax.view_init(30, -35)
    #ax.grid(False)
    #ax.set_xticks([])
    #ax.set_yticks([])
    #ax.set_zticks([])

    return data, layout#, ax#plt.show()

def new_points(e):
  points = {"x": [],
            "y": [],
            "z": []} 
  #e = np.interp(e, range(0, 15, 1), np.linspace(0, 3, 15))         
  #print(e)  
  e_corner = np.linspace(0, 3, 4)
  
  e1, e2, e3, e4 = sorted(e_corner)
  if e1 < e < e2 :
    print("case : e1 < e < e2 ")
    points["x"] = [(e-e1)/(e2-e1),              0,              0]
    points["y"] = [             0, (e-e1)/(e3-e1),              0]
    points["z"] = [             0,              0, (e-e1)/(e4-e1)]
    
    
  elif e2 < e < e3 :
    print("case : e2 < e < e3 ")
    points["x"] = [ 3-np.sqrt(2)*((e-e1)/(e4-e2))/2, (e-e1)/(e2-e1),              0,              0]
    points["y"] = [                               0, (e-e1)/(e2-e1), (e-e1)/(e3-e1),              0]
    points["z"] = [                  (e-e1)/(e4-e1),              0,              0, (e-e1)/(e4-e1)]
    #
    #points = [[(e-e1)/(e2-e1),              0, (e-e2)/(e4-e2)*np.sqrt(2)/2],
    #          [(e-e1)/(e2-e1), (e-e1)/(e3-e1),                           0],
    #          [             0, (e-e1)/(e3-e1),                           0],
    #          [             0,              0,              (e-e1)/(e4-e1)]]
    
  elif e3 < e < e4 :
    print("case : e3 < e < e4 ")
    #points["x"] = [(e-e1)/(e2-e1),              0,              0]
    points["x"] = [3-np.sqrt(2)*((e-e1)/(e4-e2))/2,                               0,              0]
    points["y"] = [                              0, 3-np.sqrt(2)*((e-e2)/(e4-e3))/2,              0]
    points["z"] = [                 (e-e1)/(e4-e1),                  (e-e1)/(e4-e1), (e-e1)/(e4-e1)]
    #points = [[(e-e1)/(e2-e1),              0,              0],
    #          [             0, (e-e1)/(e3-e1),              0],
    #          [             0,              0, (e-e1)/(e4-e1)]]                         
  
  else:
    for k in points.keys():
      points[k] = np.zeros(shape=(1,3))
  #if points:
  #  for k in points.keys() :
  #      points[k] = [ np.interp( x , points[k], np.linspace( 0, 1, len(points[k]) ) ) for x in points[k] ]

  return points

if __name__ == "__main__":    
  Data, Layout = init_fig()
  Fig = go.FigureWidget(data=Data, layout=Layout)
  
  @interact(e = (0., 4., .1))
  def update(e):
    pts = new_points(e)
  
    #print(pts)
  #Fig.scatter(xs = pts["x"], ys= pts["y"], zs= pts["z"], color="red")
    if np.shape(Fig.data)[0] > 2: 
      Fig.data = [Fig.data[0], Fig.data[1]] 
    
    trace = go.Scatter3d(
            # 4 vertices of a tetrahedron
            x= pts["x"],
            y= pts["y"],
            z= pts["z"],
            mode='markers+text',
            )
    
    Fig.add_traces(trace)

  display(wg.VBox([ Fig ]) )
  

```
